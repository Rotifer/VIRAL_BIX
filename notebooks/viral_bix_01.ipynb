```python
def get_gene_sequence_map(gb_record):
    """Extract all feature sequences from the record."""
    gene_sequence_map = {}
    parent_seq = gb_record.seq
    for feature in gb_record.features:
        print(feature.qualifiers)
        if feature.type == "gene":
            gene_name = feature.qualifiers["gene"][0]
            gene_sequence = feature.extract(parent_seq)
            gene_sequence_map[gene_name] = str(gene_sequence)
    return gene_sequence_map
```

```python
print(get_gene_sequence_map(gb_record))
```
